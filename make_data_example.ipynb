# Make a larger manifest 

```python
coverage_more_samples = pd.read_csv('data/example_mad4hatter_pipeline_output_larger/sample_coverage.txt', sep='\t')
coverage_more_samples.head()
```

```python
# SampleID	Experiment	SampleType	Column	Row
manifest = coverage_more_samples[['SampleID']].drop_duplicates()
```

```python
# Add SampleType, Batch, Column, Row
sample_type = []
plate_row = []
col = []
batch = []
i = 0
r = 1
c = 0
cols = ['A','B','C','D','E','F','G','H']
batch_name = 'run1'
for index, row in manifest.iterrows(): 
    i += 1 

    # Add SampleType 
    if 'NegativeControl' in row.SampleID: 
        sample_type.append('negative')
    elif 'PositiveControl' in row.SampleID: 
        sample_type.append('positive')
    else: 
        sample_type.append('sample')

    # Add row 
    plate_row.append(r)
    r += 1
    if r == 12: 
        r = 0 
        
    # Add col
    col.append(cols[c])
    c += 1 
    if c == (len(cols)-1) : 
        c = 0 

    # Batch 
    batch.append(batch_name)
    if i == 60: 
        batch_name = 'run2'
        c = 0 
        r = 1
```

```python
# Add SampleType, Batch, Column, Row
manifest['SampleType'] = sample_type
manifest['Batch'] = batch
manifest['Column'] = col
manifest['Row'] = plate_row
```

```python
manifest.to_csv('manifest_large_multibatch.csv')
```

# Make a 1 batch full plate example 

```python
# SampleID	Experiment	SampleType	Column	Row
manifest = coverage_more_samples[['SampleID']].drop_duplicates()
manifest = manifest.sample(96)
```

```python
# Add SampleType, Batch, Column, Row
sample_type = []
plate_row = []
col = []
r = 1
c = 0
cols = ['A','B','C','D','E','F','G','H']
batch_name = 'run1'
for index, row in manifest.iterrows(): 
    # Add SampleType 
    if 'NegativeControl' in row.SampleID: 
        sample_type.append('negative')
    elif 'PositiveControl' in row.SampleID: 
        sample_type.append('positive')
    else: 
        sample_type.append('sample')

    # Add row 
    plate_row.append(r)
    r += 1
    if r == 12: 
        r = 0 
        
    # Add col
    col.append(cols[c])
    c += 1 
    if c == (len(cols)-1) : 
        c = 0 
```

```python
# Add SampleType, Batch, Column, Row
manifest['SampleType'] = sample_type
manifest['Batch'] = batch_name
manifest['Column'] = col
manifest['Row'] = plate_row
```
